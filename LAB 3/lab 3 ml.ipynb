```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set()
```

```python
df = pd.read_csv("Smartphone_Usage_Productivity_Dataset_50000.csv")

df.head(7)
```

```python
# finding number of rows and columns

print("Shape (rows, columns): ", df.shape,"\n")

print("number of rows: ", df.shape[0])
print("number of columns: ", df.shape[1])
```

```python
df = df.dropna()
```

```python
# checking duplicate rows
df.duplicated()[df.duplicated()==True]
```

```python
# Statistical summary
df.describe(include='all')
```

```python
plt.figure(figsize=(8,5))
sns.histplot(df['Age'], bins=20, kde=True)
plt.title("Age Distribution")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.histplot(df['Daily_Phone_Hours'], bins=20, kde=True)
plt.title("Daily Phone Usage")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.scatterplot(x='Daily_Phone_Hours', y='Work_Productivity_Score', data=df)
plt.title("Phone Usage vs Productivity")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.scatterplot(x='Sleep_Hours', y='Stress_Level', data=df)
plt.title("Sleep Hours vs Stress Level")
plt.show()
```

```python
sns.boxplot(x='Device_Type', y='Work_Productivity_Score', data=df)
plt.title("Productivity by Device Type")
plt.show()
```

```python
plt.figure(figsize=(10,8))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap="coolwarm")
plt.title("Correlation Matrix")
plt.show()
```
